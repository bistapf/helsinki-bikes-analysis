```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import numpy as np
```

```python
# Import data from csv file
df = pd.read_csv('database.csv')

# Print
df
```

```python
#clean
# Replace spaces with underscores in column names
df.columns = df.columns.str.replace(' ', '_')

# Print head
df.head()
```

```python
#filter out only 2019 data for smaller starter dataset
df_2019 = df[df['departure'].str.contains('2019') ]
```

```python
df_2019.head()
```

```python
# take only every 10th entry for a smaller test dataset to use in streamlit dashboard

df_2019_small = df_2019.iloc[::10, :]

df_2019_small
            
```

```python
#filter out rides with no duration
#check for NaN values
df_2019_small_cleaned = df_2019_small.dropna()

nan_columns = df_2019_small_cleaned.columns[df_2019_small_cleaned.isna().any()].tolist()
print("Columns that contain NaNs: " +str(nan_columns))

nan_rows = df_2019_small_cleaned[df_2019_small_cleaned.isna().any(axis=1)]
print(nan_rows)
```

```python
df_2019_small_cleaned
```

```python
#drop the columns that we dont need:
df_2019_small_cleaned_thinned = df_2019_small_cleaned.drop(['departure_name', 'return_name'], axis=1)

df_2019_small_cleaned_thinned
```

```python
#add the weekday and month columns
df_2019_small_cleaned_thinned['departure'] = pd.to_datetime(df['departure'])
df_2019_small_cleaned_thinned['departure_weekday'] = df_2019_small_cleaned_thinned['departure'].dt.strftime('%A')
df_2019_small_cleaned_thinned['departure_month'] = df_2019_small_cleaned_thinned['departure'].dt.strftime('%B')
```

```python
df_2019_small_cleaned_thinned.to_csv('df_2019_small_cleaned_thinned.csv', index=False)
```
