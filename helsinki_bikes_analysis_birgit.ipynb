```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import numpy as np
```

```python
# Import data from csv file
df = pd.read_csv('data_prep/dataset_2019_only.csv.gz')

# Print
df
```

```python
#try to work with datetime

df['departure'] = pd.to_datetime(df['departure'])
# df['departure_weekday'] = df['departure'].dt.dayofweek
df['departure_weekday'] = df['departure'].dt.strftime('%A')

df.head()
```

```python
#check on number of rides per weekday
departures_per_weekday = df['departure_weekday'].value_counts()

#and plot as histogram
plt.figure(figsize=(8, 6))
departures_per_weekday.plot(kind='bar')
plt.xlabel('Weekday')
plt.ylabel('Number of departures')
plt.show()
```

```python
#same for month
df['departure_month'] = df['departure'].dt.strftime('%B')

# df.head()

#check on number of rides per weekday
departures_per_month = df['departure_month'].value_counts()

#and plot as histogram
plt.figure(figsize=(8, 6))
departures_per_month.plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Number of departures')
plt.show()
```

```python
dep_station_counts = df['departure_id'].value_counts()
df['departure_station_frequency'] = df['departure_id'].map(dep_station_counts)
df_departures=df[['departure_longitude','departure_latitude','departure_station_frequency']].copy().drop_duplicates()
sns.scatterplot(data=df_departures, x="departure_longitude", y="departure_latitude", hue="departure_station_frequency", size="departure_station_frequency", legend=True)
```

```python
arr_station_counts = df['return_id'].value_counts()
df['arrival_station_frequency'] = df['return_id'].map(arr_station_counts)
df_arrivals=df[['return_longitude','return_latitude','arrival_station_frequency']].copy().drop_duplicates()
sns.scatterplot(data=df_arrivals, x="return_longitude", y="return_latitude", hue="arrival_station_frequency", size="arrival_station_frequency", legend=True)
```

```python
df_departure = df.groupby('departure_id')
df_departure['distance_(m)'].mean()
df['average_distance'] = df['departure_id'].map(df_departure['distance_(m)'].mean())
df_meandistance=df[['departure_longitude','departure_latitude','average_distance']].copy().drop_duplicates()
sns.scatterplot(data=df_meandistance, x="departure_longitude", y="departure_latitude", hue="average_distance", size="average_distance", legend=True)
```
